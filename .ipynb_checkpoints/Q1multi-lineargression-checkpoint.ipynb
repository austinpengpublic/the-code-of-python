```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
```

```python
df.head()
```

```python
col_names = ["1) Alcohol", "2) Malic acid" ,"3) Ash" ,"4) Alcalinity of ash",  "5) Magnesium" ,"6) Total phenols" 
             ,"7) Flavanoids" ,"8) Nonflavanoid phenols", "9) Proanthocyanins" ,"10) Color intensity", "11) Hue" ,
             "12) OD280/OD315 of diluted wines" ,"13) Proline" ]
```

```python
print (col_names)
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
                names = col_names, 
                header = None)
```

```python
df.head()
```

```python
df.index
```

```python
df.index = range(1,179)
```

```python
df.index
```

```python
df.corr()
```

```python
## y data
y = np.array(df['12) OD280/OD315 of diluted wines'])
y = y.reshape(len(y),1)
print (y)
```

```python
## x data
x= np.array([df['6) Total phenols'] , df['7) Flavanoids']]) 
x = x.T
print (x)
```

```python
## model
regr = LinearRegression()
regr.fit(x,y)
y_hat = regr.predict(x)
```

```python
##result
r_square = regr.score(x,y)
print (regr.coef_)
print (regr.intercept_)
print (r_square)
```

```python
x_test = np.array([[1.3, 0.75]])
print (x_test)
regr.predict(x_test)
```

```python
from mpl_toolkits.mplot3d import Axes3D

x1_plot = np.array(df['6) Total phenols']) 
x1_plot = x1_plot.reshape(1,len(x1_plot))
x2_plot = np.array(df['7) Flavanoids'])
x2_plot = x2_plot.reshape(1,len(x2_plot))
y_plot = y_hat.reshape(1,len(y_hat))
ax = plt.subplot(projection='3d')  
ax.scatter(x1_plot, x2_plot, y_plot)
```
