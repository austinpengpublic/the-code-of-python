```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing

col_names = ["1) Alcohol", "2) Malic acid" ,"3) Ash" ,"4) Alcalinity of ash",  "5) Magnesium" ,"6) Total phenols" 
             ,"7) Flavanoids" ,"8) Nonflavanoid phenols", "9) Proanthocyanins" ,"10) Color intensity", "11) Hue" ,
             "12) OD280/OD315 of diluted wines" ,"13) Proline" ]
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
                names = col_names, 
                header = None)
df.index = range(1,179)
df.head()
```

```python
#x matrix
x= np.array([df['6) Total phenols'] , df['7) Flavanoids']]) 
x = x.T

print ("x:")
print (x)
```

```python
##normolize
nx=preprocessing.scale(x)
nx
```

```python
##cov
t=nx.T
cov= np.cov(t)
cov
```

```python
##eginvetor 
value, vector=np.linalg.eig(cov)

print("value")
print(value)

print ("vector")
print (vector)
```

```python
## 
sq=value**0.5
ve=np.diag(sq)
ve
```

```python
##A:loading
A=ve.dot(vector.T)
A
```

```python
##F: factor score
ia=np.linalg.inv(A)
F=nx.dot(ia)
F
```

```python
##communality
h1=np.sum(A[0]**2)
h2=np.sum(A[1]**2)

np.array([h1,h2])
```

```python
#uniqueness
u1=(-0.70710678)**2-h1
u2=(-0.70710678)**2-h2

np.array([u1,u2])
```
