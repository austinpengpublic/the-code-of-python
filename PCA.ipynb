```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing

col_names = ["1) Alcohol", "2) Malic acid" ,"3) Ash" ,"4) Alcalinity of ash",  "5) Magnesium" ,"6) Total phenols" 
             ,"7) Flavanoids" ,"8) Nonflavanoid phenols", "9) Proanthocyanins" ,"10) Color intensity", "11) Hue" ,
             "12) OD280/OD315 of diluted wines" ,"13) Proline" ]
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
                names = col_names, 
                header = None)
df.index = range(1,179)
df.head()
```

```python
#x matrix
x= np.array([df['6) Total phenols'] , df['7) Flavanoids']]) 
x = x.T

print ("x:")
print (x)
```

```python
##normolize
nx=preprocessing.scale(x)
nx
```

```python
##cov
t=nx.T
cov= np.cov(t)
cov
```

```python
##eginvetor 
value, vector=np.linalg.eig(cov)

print("value")
print(value)

print ("vector/loading matix")
print (vector)
```

```python
##score matrix
nx.dot(vector)
```

```python
PC=["pc1","pc2"]
plt.bar(PC,value)
```

```python
sv=np.sum(value)
p=value/sv
cum=np.cumsum(p)
plt.plot(PC,cum, color='r')
```
